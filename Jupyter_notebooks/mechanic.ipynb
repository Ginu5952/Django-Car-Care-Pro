                                    MECHANIC

1) Add Mechanic

- `api/v1/mechanic/`
    - `POST`: Add a mechanic

        - ** Headers**
            - Authorization Token


        - Request Body

        ```json
            {
            "user": 6,
            "years_of_experience": 5,
            "preferred_job_types": "Full-time",
            "address": "123 Mechanic Street, Mechanic City",
            "gender": "M",
            "contact_number": "123456789",
            "profile_image": null,
            "work_days": ["Monday", "Wednesday", "Friday"]
            }
        ```    

        - Response Body

            - Status Code : `201 Created`

        ```json
            {
                "id": 1,
                "user": 6,
                "user_full_name": "jestin joy",
                "years_of_experience": 5,
                "preferred_job_types": "Full-time",
                "address": "123 Mechanic Street, Mechanic City",
                "gender": "M",
                "contact_number": "123456789",
                "profile_image": null,
                "work_days": [
                    "Monday",
                    "Wednesday",
                    "Friday"
                ]
            }
        ```    

        - Response Body

            - Status Code: `400 Bad Request`

        ```json
            {
            "user": [
                "Invalid pk \"11\" - object does not exist."
            ]
            }
        ```    


        -  Response Body

            - Status Code: `401 Unauthorized`

        ```json
            {
            "detail": "Invalid token."
            }    
        ```   

2) Add Mechanic's Services price per hour details

- `api/v1/mechanic/price/`
    - `POST` : Add hourly rate per service

    - Request Body

    ```json
    {
    "mechanic": 1,
    "service_name": "Oil Change",
    "hourly_rate": "160.00",
    "currency": "Euro"
    }
    ```

    - Response Body
    
        - Status Code : `201 Created`

    ```json
    {
    "id": 4,
    "mechanic_name": "jestin joy",
    "service_name": "Oil Change",
    "hourly_rate": "160.00",
    "currency": "Euro"
    }  
    ```

    - Response Body
        - Status Code : `400 Bad Request`

    ```json
    {
    "mechanic": "Invalid mechanic ID."
    }
    ```
