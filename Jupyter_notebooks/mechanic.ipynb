12) Retrieve all customer feedbacks

    - `/api/v1/customer/feedbacks/`
        - `GET`
            - **Header**
                - Authentication `Token`
                - Accept `application/json`

            - RESPONSE BODY

                Status Code: `200 OK`

                ```json
                    {
                        [
                            {
                                "feedback_id": 1,
                                "customer_id": 123,
                                "service_request_id": 789,
                                "rating": 5,
                                "comments": "Excellent service! The mechanic was very professional and fixed the issue promptly.",
                                "date_of_feedback": "2024-08-15"
                            },
                            {
                                "feedback_id": 2,
                                "customer_id": 124,
                                "service_request_id": 790,
                                "rating": 4,
                                "comments": "Good service, but there was a delay in getting the parts needed for repair.",
                                "date_of_feedback": "2024-08-16"
                            }
                            // More feedback entries
                        ]
                }
                ```

                - RESPONSE BODY

                    Status Code: `204 No Content`

                        ```json
                            {
                                "message": "No feedbacks available.",
                                "error": "No_Content"
                            }
                        ```   

                    Status Code: `500 Internal Server Error`

                    ```json
                        {
                        "error": "Internal_Server_Error",
                        "message": "An unexpected error occurred while retrieving feedbacks."
                        }
                    ```
