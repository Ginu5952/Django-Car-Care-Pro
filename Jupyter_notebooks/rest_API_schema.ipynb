# CarCarPro API Schema

1) customer sign up

- `/api/v1/customer/signup/`
    - `POST`
        - REQUEST BODY

               
                    {             
                        
                        "username": {
                            "type": "str",
                            "value": "kevin",
                            "required": true
                        },
                        "first_name": {
                            "type": "str",
                            "value": "kenz",
                            "required": true
                        },
                        "last_name": {
                            "type": "str",
                            "value": "eyong",
                            "required": false
                        },
                        "address": {
                            "type": "str",
                            "value": "123 Main St, Hamburg, Germany",
                            "required": false
                        },
                        "email": {
                            "type": "str",
                            "value": "kevin@gmail.com",
                            "required": false
                        },
                        "gender": {
                            "type": "str",
                            "value": "male",
                            "required": false
                        },
                        "contact_number": {
                            "type": "str",
                            "value": "+49 123 456 7890",
                            "required": false
                        },
                        "profile_image": {
                            "type": "str",
                            "value": "http://example.com/images/profile.jpg",
                            "required": false
                        },
                        "date_of_join": {
                            "type": "str",
                            "value": "2024-08-14",
                            "required": false
                        }
                    }
               

        - RESPONSE BODY

            - Status code: `201 Created`

            ```json
                {
                    "message": "Successfully sign up.",
                    "username": "kevin",
                    "customer_id": 123
                }
            ```
        - RESPONSE BODY

            - Status code: `400 Bad Request`

            ```json
                {
                    "message": "Signup failed. Please check the input fields.",
                    "errors": {
                        "username": "This username is already taken.",
                        "email": "A valid email address is required."
                    }
                }
            ```



2) Customer Login

- `api/v1/customer/login/`
    - `POST`
        - REQUEST BODY

            ```json
                {
                    "username": {
                        "type": "str",
                        "value": "kevin",
                        "required": true
                    },
                    "password": {
                        "type": "str",
                        "value": "*********",
                        "required": true
                    }
                }

        - RESPONSE BODY

            - Status code: `200 OK`

            ```json
                {
                "message": "Login successful. Welcome to Car Care Pro!",
                "username": "kevin",
                "token": "eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.e30.4Qy5IR-FMJND7e6QSkZr6VvC0oQhQhJ-Y3SoDLMeGbU"
                }
            ```   

        - RESPONSE BODY

            - Status Code: `401 Unauthorized`     

            ```json
                {
                "message": "Login failed. Invalid username or password."
                }

5) Customer Service Request

    - `/api/v1/service-request/`
        - `POST`
            - REQUEST BODY

                ```json
                    {
                    "category": {
                        "type": "str",
                        "value": "four_wheeler",
                        "required": true,
                        "choices": ["two_wheeler_with_gear", "two_wheeler_without_gear", "three_wheeler", "four_wheeler"]
                    },
                    "vehicle_no": {
                        "type": "str",
                        "value": "MH12AB1234",
                        "required": true
                    },
                    "vehicle_name": {
                        "type": "str",
                        "value": "Honda City",
                        "required": true
                    },
                    "vehicle_model": {
                        "type": "str",
                        "value": "2018",
                        "required": true
                    },
                    "vehicle_brand": {
                        "type": "str",
                        "value": "Honda",
                        "required": true
                    },
                    "problem_description": {
                        "type": "str",
                        "value": "Engine overheating issue.",
                        "required": true
                    },
                    "date_of_request": {
                        "type": "str",
                        "value": "2024-08-14",
                        "required": true
                    },
                    "service_needed": {
                        "type": "int",
                        "value": 1,
                        "required": true
                    },
                    "customer_id": {
                        "type": "int",
                        "value": 123,
                        "required": true
                    },
                    "mechanic_id": {
                        "type": "int",
                        "value": 456,
                        "required": false
                    }
                }
                ```

            - RESPONSE BODY

                Status Code: `201 Created`

                ```json
                    {
                    "message": "Service request created successfully.",
                    "request_id": 789
                    }
                ```    

            - RESPONSE BODY

                Status Code: `400 Bad Request`

                ```json
                    {
                        "message": "Service request creation failed. Please check the input fields.",
                        "errors": {
                            "category": "This field is required and must be one of the specified choices.",
                            "vehicle_no": "This field is required.",
                            "vehicle_name": "This field is required.",
                            "vehicle_model": "This field is required.",
                            "vehicle_brand": "This field is required.",
                            "problem_description": "This field is required.",
                            "date_of_request": "This field is required and must be a valid date.",
                            "service_needed": "Service needed must be a valid service ID.",
                            "customer": "Customer ID is required and must be a valid integer.",
                            "mechanic": "Mechanic ID must be a valid integer if provided."
                        }
                    }
                ```

6) Fetch Customer Service Request

- `/api/v1/customer/{customer_id}/`
    - `GET`

        - RESPONSE BODY

            Status Code: `200 OK`

            ```json
                {
                "customer_id": 123,
                "service_requests": [
                    {
                        "request_id": 789,
                        "category": "four_wheeler",
                        "vehicle_no": "MH12AB1234",
                        "vehicle_name": "Honda City",
                        "vehicle_model": "2018",
                        "vehicle_brand": "Honda",
                        "problem_description": "Engine overheating issue.",
                        "date_of_request": "2024-08-14",
                        "service_needed": 1,  // ID of the service
                        "mechanic": 456       
                    },
                    {
                        "request_id": 790,
                        "category": "two_wheeler_with_gear",
                        "vehicle_no": "MH12CD5678",
                        "vehicle_name": "Yamaha FZ",
                        "vehicle_model": "2020",
                        "vehicle_brand": "Yamaha",
                        "problem_description": "Brake issue.",
                        "date_of_request": "2024-08-15",
                        "service_needed": 2, 
                        "mechanic": 457       
                    }
                
                ]
            }
            ```

        - RESPONSE BODY

            Status Code: `404 Not Found`
            ```json
                {
                    "message": "Customer not found.",
                    "error": "The customer with ID 123 does not exist."
                }
            ```    

7) Get All Service Requests

- `/api/v1/service-requests/`
    - `GET`
        - RESPONSE BODY 

            Status Code: `200 OK`

            ```json
                {

                "service_requests": [
                    {
                        "request_id": 789,
                        "category": "four_wheeler",
                        "vehicle_no": "MH12AB1234",
                        "vehicle_name": "Honda City",
                        "vehicle_model": "2018",
                        "vehicle_brand": "Honda",
                        "problem_description": "Engine overheating issue.",
                        "date_of_request": "2024-08-14",
                        "service_needed": 1,  
                        "customer_id": 123,     
                        "mechanic_id": 456       
                    },
                    // More service requests
                ]
            }
            ```  

        - RESPONSE BODY

            Status Code: 403 Forbidden

            ```json
                {
                "message": "Permission denied.",
                "error": "You do not have permission to access service requests for this customer."
                }
            ```
